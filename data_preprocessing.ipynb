```python
# Perform an inner join
merged_data = pd.merge(players_data, players_stat, left_on='player_id', right_on= 'id', how= 'inner')

# Display the first few rows of the merged dataset
merged_data["minutes_played"].head()
```

### Change the minutes_played into integer type

```python
# Replace hyphens with NaN and convert to float
# inplace transform
# Modified function to handle dashes and missing values in the minutes_played column
def clean_minutes_played(value):
    if pd.isna(value) or value == '-':
        return None
    return int(value.replace("'", "").replace(".", ""))

# Apply the modified function to the minutes_played column
merged_data['minutes_played'] = merged_data['minutes_played'].apply(clean_minutes_played)

```

```python
merged_data["minutes_played"].head()
```

```python
merged_data.columns
```

### Extract only age, remove birthday

```python
# Extract age, remove birthday
# Extract age from the date_of_birth column
merged_data['age'] = merged_data['date_of_birth'].str.extract(r'\((\d+)\)').astype(float)

# merged_data[['date_of_birth', 'age']].head()
merged_data.head()
```

```python
# Then drop date_of_birth column
merged_data.drop(columns= ["date_of_birth"], inplace= True)
print(merged_data.head())
```

```python
merged_data.head()
```

### We have playerID, name is not neccesary for computer

```python
merged_data.drop(columns= ['given_name', 'full_name', "name", "players_link", "id"], inplace= True)
```
